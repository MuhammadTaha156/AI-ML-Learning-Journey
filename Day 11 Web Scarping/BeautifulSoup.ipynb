# Practice Web Sracping using BeautifulSoup

```python
with open("scraped_data/data2.html","r") as f:
    html_content=f.read()
```

```python
html_content
print(html_content)
```

```python
soup=bs(html_content,"lxml")
soup
```

```python
table=soup.find("table",class_="table")
table
```

```python
Team_details=[]
for tr in table.find_all("tr",class_="team"):
    # Extract Team Name
    name=tr.find("td",class_="name")
    team_name=name.get_text(strip=True)
    
    # Extract Wins
    win_td=tr.find("td",class_="wins")
    wins=win_td.get_text(strip=True)

    # Extract Losses
    loss_td=tr.find("td",class_="losses")
    losses=loss_td.get_text(strip=True)

    Team_details.append([team_name,wins,losses])
```

```python
print(Team_details)
```

```python
df=pd.DataFrame(Team_details,columns=["Team Name","Wins","Losses"])
df
```

```python
df.to_csv("team deatils.csv",index=False)
```

```python
with open("scraped_data/data6.html","r") as f:
    html_content=f.read()
```

```python
html_content
```

```python
soup=bs(html_content,'lxml')
```

```python
soup
```

```python
all_tr=soup.find_all('tr',class_="team")
```

```python
Hockey_Team=[]
for tr in all_tr:
    Team_Name=tr.findAll("td",class_="name")[0].get_text(strip=True)
    Year=tr.findAll("td",class_="year")[0].get_text(strip=True)
    Wins=tr.findAll("td",class_="wins")[0].get_text(strip=True)
    losses=tr.findAll("td",class_="losses")[0].get_text(strip=True)

    Hockey_Team.append([Team_Name,Year,Wins,losses])
```

```python
df=pd.DataFrame(Hockey_Team,columns=["Team Name","Year","Wins","Losses"])
```

```python
df
```

```python
df.to_csv("Hockey Team Info",index=False)
```

```python
with open("scraped_data/data10.html","r") as f:
    html_content=f.read()
```

```python
soup=bs(html_content,'lxml')
```

```python
soup
```

```python
all_h4 = soup.find_all("h4")
```

```python
all_h4
```

```python
Real_World_Challenges=[]
for h in all_h4:
    Heading=h.find_all("a")[0].get_text(strip=True)
    para=h.findAllNext("p")[0].get_text(strip=True)
    Real_World_Challenges.append([Heading,para])
```

```python
Real_World_Challenges
```

```python
df=pd.DataFrame(Real_World_Challenges,columns=["Heading","Para"])
df
```
