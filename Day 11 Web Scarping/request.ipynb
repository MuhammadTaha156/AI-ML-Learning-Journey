```python
import requests as r
import pandas as pd
```

```python
URL="https://stephen-king-api.onrender.com/api/books"
```

```python
respones=r.get(URL)
respones
```

```python
json_Data=respones.json()
json_Data
```

```python
df=pd.json_normalize(json_Data["data"])
df
```

```python
df.columns
```

```python
df[["id","Title","Year","ISBN"]]
```

# Web Scarping

```python
url="https://www.scrapethissite.com/pages/simple/"
```

```python
headers={
    "User-Agent":"Mozilla/5.0"
}
res=r.get(url,headers=headers)
res
```

```python
res.status_code
```

```python
res.headers
```

```python
res.content
```

```python
res.text
```

```python
if res.status_code==200:
    print(res.text)
    print(res.content)
    print(res.headers)
```

```python
with open("scraped_data/data1.html","w") as f:
    f.write(res.text)
```
